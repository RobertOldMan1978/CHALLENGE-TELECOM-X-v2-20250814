```python
# Creamos un módulo reutilizable con el estándar visual en tonos azules
from matplotlib import rcParams
from matplotlib.ticker import StrMethodFormatter
import matplotlib as mpl

BLUE_PALETTE = ["#0D47A1", "#1565C0", "#1976D2", "#1E88E5", "#42A5F5", "#90CAF9"]
BG_PAPER = "#F5F5F5"
GRID_LIGHT = "#E0E0E0"

def set_estandar_azul(base_fontsize: int = 12, usar_grid: bool = True):
    """
    Aplica un estilo global en tonos azules para matplotlib.
    - base_fontsize: tamaño base de fuente
    - usar_grid: activa o desactiva la cuadrícula ligera
    Devuelve la paleta usada.
    """
    rcParams.update({
        "figure.facecolor": BG_PAPER,
        "axes.facecolor": "#FFFFFF",
        "axes.edgecolor": GRID_LIGHT,
        "axes.spines.top": False,
        "axes.spines.right": False,
        "axes.titleweight": "bold",
        "axes.titlesize": base_fontsize + 2,
        "axes.labelsize": base_fontsize,
        "axes.labelweight": "bold",
        "xtick.labelsize": base_fontsize - 1,
        "ytick.labelsize": base_fontsize - 1,
        "grid.color": GRID_LIGHT,
        "grid.linestyle": "-",
        "grid.linewidth": 0.6,
        "legend.frameon": False,
    })
    if usar_grid:
        rcParams.update({"axes.grid": True})
    else:
        rcParams.update({"axes.grid": False})
    # Ciclo de colores en tonos azules
    rcParams["axes.prop_cycle"] = mpl.cycler(color=BLUE_PALETTE)
    return BLUE_PALETTE

def formatear_eje_y_miles(ax):
    """Formatea el eje Y con separador de miles sin decimales."""
    ax.yaxis.set_major_formatter(StrMethodFormatter("{x:,.0f}"))
    return ax

def reset_estandar():
    """Restaura la configuración por defecto de matplotlib."""
    rcParams.update(mpl.rcParamsDefault)

```

## EVALUACION TRATAMIENTO DESBALANCEO

1️⃣ SMOTE (Synthetic Minority Over-sampling Technique)

Cómo funciona:

    Genera nuevos ejemplos sintéticos de la clase minoritaria interpolando registros existentes.
    No borra datos de la clase mayoritaria.
    Mantiene toda la información disponible y mejora el entrenamiento en clases con pocos casos.

Ventajas:

    Aprovecha todos los datos originales.
    Reduce el riesgo de sobreajuste que produce el simple duplicado de registros.
    Útil cuando la clase minoritaria es muy pequeña.

Desventajas:

    Genera datos artificiales que pueden introducir ruido si las variables son muy dispersas.
    Menos efectivo si la clase minoritaria está muy fragmentada en el espacio de características.

2️⃣ Undersampling

Cómo funciona:

    Elimina registros de la clase mayoritaria hasta equilibrar con la minoritaria.
    Entrena con menos datos, pero balanceados.

Ventajas:

    Reduce drásticamente el tiempo de entrenamiento.
    Simple y fácil de implementar.

Desventajas:

    Pérdida de información potencialmente valiosa.
    Si la clase mayoritaria tiene patrones importantes, pueden perderse y afectar la generalización.

```python
# (opcional) instalar si aún no lo tienes en tu entorno:
# !pip install imbalanced-learn

from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler

# 1) Separar X e y
X = datos_final.drop(columns='cliente.abandono')
y = datos_final['cliente.abandono'].astype(int)  # asegúrate de 0/1

# 2) SMOTE (oversampling sintético)
smote = SMOTE(random_state=42)
X_smote, y_smote = smote.fit_resample(X, y)

# 3) Undersampling aleatorio
under = RandomUnderSampler(random_state=42)
X_under, y_under = under.fit_resample(X, y)

# 4) Resumen rápido
def resumen(nombre, y_vec):
    vc = y_vec.value_counts().sort_index()
    total = len(y_vec)
    p0 = (vc.get(0,0)/total)*100
    p1 = (vc.get(1,0)/total)*100
    print(f"{nombre:>12} | Total: {total:5d} | 0: {vc.get(0,0):5d} ({p0:6.2f}%)  | 1: {vc.get(1,0):5d} ({p1:6.2f}%)")

print("\nDistribución de clases")
resumen("Original", y)
resumen("SMOTE", y_smote)
resumen("Under", y_under)

# (opcional) guardar para usar luego
# X_smote.to_csv("X_smote.csv", index=False); y_smote.to_csv("y_smote.csv", index=False)
# X_under.to_csv("X_under.csv", index=False); y_under.to_csv("y_under.csv", index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Aplicar estándar azul
set_estandar_azul(base_fontsize=11, usar_grid=False)

# Calcular matriz de correlación con todas las columnas numéricas
corr_matrix = datos_final.corr()

# Crear heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, cmap="Blues", annot=False, center=0, cbar_kws={"shrink": 0.8})
plt.title("Matriz de Correlación - Todas las Variables Numéricas", fontsize=14, fontweight='bold')
plt.show()

# Detectar pares con alta correlación (>= 0.85 o <= -0.85, excluyendo la diagonal)
umbral = 0.85
alta_corr = []
for col in corr_matrix.columns:
    for idx in corr_matrix.index:
        if col != idx and abs(corr_matrix.loc[idx, col]) >= umbral:
            alta_corr.append((idx, col, corr_matrix.loc[idx, col]))

# Eliminar duplicados (pares repetidos en orden inverso)
pares_unicos = []
vistos = set()
for a, b, c in alta_corr:
    if (b, a) not in vistos:
        pares_unicos.append((a, b, c))
        vistos.add((a, b))

pares_unicos[:10]  # mostramos solo los 10 primeros pares con alta correlación

```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

def reporte_metricas(y_true, y_pred, y_proba=None, titulo="Métricas"):
    acc  = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, zero_division=0)
    rec  = recall_score(y_true, y_pred, zero_division=0)
    f1   = f1_score(y_true, y_pred, zero_division=0)

    print(f"=== {titulo} ===")
    print(f"Accuracy : {acc:.4f}")
    print(f"Precision: {prec:.4f}")
    print(f"Recall   : {rec:.4f}")
    print(f"F1       : {f1:.4f}")
    if y_proba is not None:
        try:
            auc = roc_auc_score(y_true, y_proba)
            print(f"ROC AUC  : {auc:.4f}")
        except Exception as e:
            print("ROC AUC  : no disponible (", e, ")")
```

## CORREGIR UMBRAL DE DECISION

---

### **Qué es el umbral de decisión**

* Cuando un modelo de clasificación como la **regresión logística** predice, no da directamente “Activo” o “Cancelado”, sino **probabilidades** (por ejemplo, 0.23, 0.61, 0.85…).
* Por defecto, **scikit-learn** clasifica como “Cancelado” si la probabilidad ≥ **0.5** y como “Activo” si es < 0.5.
* Ese 0.5 es el **umbral** y no siempre es el mejor valor para tu objetivo.

---

### **Por qué ajustarlo en nuestro caso**

* Nuestro objetivo principal es **detectar la mayor cantidad posible de clientes que van a cancelar** (maximizar **recall**), porque:

  * Perder un cliente y no detectarlo a tiempo es costoso.
  * Un **falso negativo** (no marcar a un cliente en riesgo que sí cancela) es más grave que un falso positivo.
* Actualmente con `class_weight=balanced` tenemos **recall \~80%**, pero quizá podríamos subirlo al **85–90%** si bajamos el umbral, sacrificando algo de precisión.

---

### **Cómo se hace**

* Barrer valores de umbral entre 0 y 1 (ejemplo: 0.1, 0.2, …, 0.9).
* Calcular métricas (precision, recall, F1, F2) para cada umbral.
* Elegir el umbral que mejor equilibre lo que nos importa (ej. recall alto sin que la precisión caiga demasiado).

---

### **Recomendación para este proyecto**

* **Sí ajustar el umbral antes de pasar a otros modelos**, porque:

  1. Nos dará un baseline **realista** con recall optimizado, contra el cual comparar modelos complejos.
  2. Permite afinar la estrategia de negocio antes de buscar mejoras con algoritmos más potentes.
  3. Así aseguramos que, al comparar dos modelos futuros (por ejemplo, un árbol y un GBDT), ambos tengan umbrales ajustados y no queden en desventaja por usar el 0.5 por defecto.




📌 **Propuesta concreta**

1. Usar la **regresión logística con `class_weight=balanced`** y hacer un barrido de umbral.
2. Evaluar con **F2** (que penaliza más los falsos negativos) y recall, y guardar ese valor como **umbral óptimo baseline**.
3. Aplicar ese mismo proceso a los dos modelos predictivos que haremos después para comparar métricas en igualdad de condiciones.

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import (
    confusion_matrix,
    ConfusionMatrixDisplay,
    classification_report,
    precision_recall_curve
)

# Ya tienes tu modelo entrenado:
# logit_bal = LogisticRegression(max_iter=200, n_jobs=-1, class_weight="balanced", random_state=42)
# logit_bal.fit(X_train, y_train)

# Probabilidades en test para la clase positiva (Cancelado=1)
y_true   = y_test
y_scores = y_proba_logit_bal  # ya calculado: logit_bal.predict_proba(X_test)[:, 1]
```

```python
# Curva Precisión–Recall a partir de y_true e y_scores
precision, recall, thresholds = precision_recall_curve(y_true, y_scores)

# Alinear arrays (thresholds tiene N-1 puntos)
precision_th  = precision[:-1]
recall_th     = recall[:-1]
thresholds_th = thresholds

# F2 (beta=2) para penalizar más falsos negativos
beta = 2.0
f2_scores = (1 + beta**2) * (precision_th * recall_th) / (beta**2 * precision_th + recall_th + 1e-12)

# Reglas de negocio (ajusta si lo necesitas)
RECALL_MIN = 0.85
PREC_MIN   = 0.40

viables = np.where((recall_th >= RECALL_MIN) & (precision_th >= PREC_MIN))[0]
if len(viables) == 0:
    viables = np.where(recall_th >= RECALL_MIN)[0]

if len(viables) == 0:
    idx_opt = int(np.argmax(f2_scores))  # mejor F2 global
else:
    idx_opt = int(viables[np.argmax(f2_scores[viables])])

thr_opt = float(thresholds_th[idx_opt])

print("=== Selección de umbral ===")
print(f"Umbral óptimo: {thr_opt:.4f}")
print(f"Precision@thr: {precision_th[idx_opt]:.4f} | Recall@thr: {recall_th[idx_opt]:{'.4f'}} | F2@thr: {f2_scores[idx_opt]:.4f}")
```

##UTILIDADES MINIMAS RECALL 0.85 PREC 0.4

```python
import numpy as np
from sklearn.metrics import precision_recall_curve, confusion_matrix, classification_report

# Selección de umbral con F2 priorizando recall (puedes ajustar RECALL_MIN y PREC_MIN)
def seleccionar_umbral(y_true, y_scores, beta=2.0, RECALL_MIN=0.85, PREC_MIN=0.40):
    precision, recall, thresholds = precision_recall_curve(y_true, y_scores)
    # Alinear arrays (thresholds tiene N-1 puntos)
    precision_th, recall_th, thresholds_th = precision[:-1], recall[:-1], thresholds
    fbeta = (1 + beta**2) * (precision_th * recall_th) / (beta**2 * precision_th + recall_th + 1e-12)

    viables = np.where((recall_th >= RECALL_MIN) & (precision_th >= PREC_MIN))[0]
    if len(viables) == 0:
        viables = np.where(recall_th >= RECALL_MIN)[0]
    if len(viables) == 0:
        idx_opt = int(np.argmax(fbeta))
    else:
        idx_opt = int(viables[np.argmax(fbeta[viables])])

    thr_opt = float(thresholds_th[idx_opt])
    return thr_opt, precision_th[idx_opt], recall_th[idx_opt], fbeta[idx_opt]
```

### TABLA DE COMPARACION EXTENDIDA

```python
import pandas as pd
from sklearn.metrics import confusion_matrix

# Función para obtener métricas completas
def resumen_modelo(nombre, y_true, y_scores, thr):
    y_pred = (y_scores >= thr).astype(int)
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    precision = tp / (tp + fp + 1e-12)
    recall    = tp / (tp + fn + 1e-12)
    beta = 2.0
    f2 = (1 + beta**2) * (precision * recall) / (beta**2 * precision + recall + 1e-12)
    return {
        "Modelo": nombre,
        "Umbral": round(thr, 3),
        "TN": tn, "FP": fp, "FN": fn, "TP": tp,
        "Precisión": round(precision, 4),
        "Recall": round(recall, 4),
        "F2": round(f2, 4)
    }

# Resumen para cada modelo
tabla_comparativa = [
    resumen_modelo("LogReg (balanced)", y_test, y_proba_logit_bal, 0.334),
    resumen_modelo("Decision Tree (d=5)", y_test, y_proba_tree, 0.331),
    resumen_modelo("Random Forest (d=5)", y_test, y_proba_rf, 0.367)
]

df_comparativa = pd.DataFrame(tabla_comparativa)
print(df_comparativa)
```

### AJUSTE DE MODELO

🎯 Estrategia de tuning

Mantener: class_weight='balanced_subsample' (ya probado para manejar el desbalance).

Probar variaciones en:

n_estimators: número de árboles (200, 300, 400).

max_depth: limitar profundidad para evitar sobreajuste (4, 5, 6, 8).

min_samples_leaf: mínimo de muestras en una hoja (1, 2, 5) para suavizar la predicción.

Evaluar: usar nuestro proceso de ajuste de umbral para cada configuración y quedarnos con la que dé mayor F2 cumpliendo recall ≥ 0.85 y precisión ≥ 0.40.

```python
from sklearn.ensemble import RandomForestClassifier
from itertools import product

# Rango de parámetros
n_estimators_list = [200, 300, 400]
max_depth_list = [4, 5, 6, 8]
min_samples_leaf_list = [1, 2, 5]

mejores_resultados = []

for n_estimators, max_depth, min_samples_leaf in product(n_estimators_list, max_depth_list, min_samples_leaf_list):

    rf = RandomForestClassifier(
        n_estimators=n_estimators,
        max_depth=max_depth,
        min_samples_leaf=min_samples_leaf,
        class_weight='balanced_subsample',
        random_state=42,
        n_jobs=-1
    )

    rf.fit(X_train, y_train)
    y_proba_rf = rf.predict_proba(X_test)[:, 1]

    thr_rf, prec_rf, rec_rf, f2_rf = seleccionar_umbral(
        y_test, y_proba_rf, beta=2.0, RECALL_MIN=0.85, PREC_MIN=0.40
    )

    mejores_resultados.append({
        "n_estimators": n_estimators,
        "max_depth": max_depth,
        "min_samples_leaf": min_samples_leaf,
        "Umbral": round(thr_rf, 3),
        "Precisión": round(prec_rf, 4),
        "Recall": round(rec_rf, 4),
        "F2": round(f2_rf, 4)
    })

# Ordenar por F2 descendente
import pandas as pd
df_tuning = pd.DataFrame(mejores_resultados).sort_values(by="F2", ascending=False)
print(df_tuning.head(10))
```

🔍 Interpretación rápida

Mejor F2 → Configuración con n_estimators=200, max_depth=8, min_samples_leaf=5.

Recall 91.18% (cumple el objetivo ≥ 0.85).

Precisión 45.47%, la más alta del top.

Umbral óptimo 0.326 (bajo, lo que ayuda a mantener recall alto).

Comparado con el segundo lugar, gana por mayor precisión con casi el mismo recall.

##  PRUEBAS MODELOS KNN Y SVM

🔹 Consideraciones técnicas

KNN

Sensible a escalado → necesitamos escalar las variables antes (StandardScaler o MinMaxScaler).

Parámetros clave a probar: n_neighbors, weights ("uniform" vs "distance").

SVM

Para probabilidades, usar SVC(probability=True).

Suele ser más costoso computacionalmente en datasets grandes, pero factible con tus datos.

Escalado de variables también necesario.

### KNN (con escalado) + ajuste de umbral

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np
import pandas as pd

# ===== KNN =====
knn_pipe = Pipeline(steps=[
    ("scaler", StandardScaler()),
    ("knn", KNeighborsClassifier(n_neighbors=21, weights="distance"))
])
knn_pipe.fit(X_train, y_train)

# Probabilidades clase positiva (Cancelado=1)
y_proba_knn = knn_pipe.predict_proba(X_test)[:, 1]

# Ajuste de umbral por F2 priorizando recall
thr_knn, prec_knn, rec_knn, f2_knn = seleccionar_umbral(
    y_test, y_proba_knn, beta=2.0, RECALL_MIN=0.85, PREC_MIN=0.40
)

# Predicciones con umbral óptimo
y_pred_knn_opt = (y_proba_knn >= thr_knn).astype(int)

print("=== KNN (umbral óptimo) ===")
print(f"Umbral: {thr_knn:.3f} | Precisión: {prec_knn:.4f} | Recall: {rec_knn:.4f} | F2: {f2_knn:.4f}")
print(classification_report(y_test, y_pred_knn_opt, target_names=["Activo","Cancelado"]))
```

### SVM (con escalado, class_weight='balanced') + ajuste de umbral

### GRAFICO PRECISION-RECALL CON UMBRAL OPTIMO MARCADO

```python
from sklearn.metrics import precision_recall_curve
import numpy as np

precision, recall, thresholds = precision_recall_curve(y_test, y_proba_rf_final)
precision_th = precision[:-1]
recall_th = recall[:-1]
thresholds_th = thresholds

# F2 para todos los thresholds
beta = 2.0
f2_all = (1 + beta**2) * (precision_th * recall_th) / (beta**2 * precision_th + recall_th + 1e-12)

fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(thresholds_th, recall_th, label="Recall", color="tab:blue")
ax.plot(thresholds_th, precision_th, label="Precisión", color="tab:orange")
ax.plot(thresholds_th, f2_all, label="F2", color="tab:green")

# Umbral óptimo
ax.axvline(thr_final, linestyle="--", color="gray", label=f"Umbral óptimo = {thr_final}")
idx_opt_plot = np.argmin(np.abs(thresholds_th - thr_final))
ax.scatter(thr_final, recall_th[idx_opt_plot], color="tab:blue", s=60)
ax.scatter(thr_final, precision_th[idx_opt_plot], color="tab:orange", s=60)
ax.scatter(thr_final, f2_all[idx_opt_plot], color="tab:green", s=60)

# Etiquetas
ax.text(thr_final, recall_th[idx_opt_plot]+0.02, f"Recall={recall_th[idx_opt_plot]:.2f}", color="tab:blue")
ax.text(thr_final, precision_th[idx_opt_plot]-0.07, f"Prec={precision_th[idx_opt_plot]:.2f}", color="tab:orange")
ax.text(thr_final, f2_all[idx_opt_plot]+0.02, f"F2={f2_all[idx_opt_plot]:.2f}", color="tab:green")

ax.set_xlabel("Umbral")
ax.set_ylabel("Métrica")
ax.set_title("Precision / Recall / F2 vs. Umbral (Random Forest optimizado)")
ax.legend()
plt.tight_layout()
plt.show()
```
